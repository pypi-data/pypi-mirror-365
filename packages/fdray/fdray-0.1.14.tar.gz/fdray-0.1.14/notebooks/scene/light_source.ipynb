```python
# #light-source-example
from fdray import *

Scene(
    Camera(0, 10, look_at=(0, 0, 0.5), view_scale=1.5, distance=3),
    LightSource((0, 1, 4), "white", from_camera=False),
    Sphere("z", 0.4, Color((0.5, 0.7, 0.7))),
    Box(-1, 1, Color("gold")).translate(0, 0, -1),
).render(200, 200)
```
