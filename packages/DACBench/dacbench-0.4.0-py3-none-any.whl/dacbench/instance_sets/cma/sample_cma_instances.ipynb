```python
from __future__ import annotations

import numpy as np
```

```python
active = [True, False]
elitist = [True, False]
orthogonal = [True, False]
sequential = [True, False]
threshold_convergence = [True, False]
step_size_adaption = ["csa", "tpa", "msr", "xnes", "m-xnes", "lp-xnes", "psr"]
mirrored = ["None", "mirrored", "mirrored pairwise"]
base_sampler = ["gaussian", "sobol", "halton"]
weights_option = ["default", "equal", "1/2^lambda"]
local_restart = ["None", "IPOP", "BIPOP"]
bound_correction = ["None", "saturate", "unif_resample", "COTN", "toroidal", "mirror"]
```

```python
def save_cma_instances(
    filename,
    n_instances=100,
    dims=None,
    function_names=None,
    iids=None,
):
    if function_names is None:
        function_names = [
            "BentCigar",
            "Discus",
            "Ellipsoid",
            "Katsuura",
            "Rastrigin",
            "Rosenbrock",
            "Schaffers",
            "Schwefel",
            "Sphere",
            "Weierstrass",
        ]
    if dims is None:
        dims = [10]
    rng = np.random.default_rng()
    if iids is None:
        iids = np.zeros(n_instances)

    with open(filename, "a") as f:
        id_string = "ID,dim,func_name,iid,active,elitist,orthogonal,sequential,threshold_convergence,step_size_adaptation,mirrored,base_sampler,weights_option,local_restart,bound_correction\n"
        f.write(id_string)
        for i in range(n_instances):
            inst_string = f"{i},{rng.choice(dims)},{rng.choice(function_names)},{rng.choice(iids)},{rng.choice(active)},{rng.choice(elitist)},{rng.choice(orthogonal)},{rng.choice(sequential)},{rng.choice(threshold_convergence)},{rng.choice(step_size_adaption)},{rng.choice(mirrored)},{rng.choice(base_sampler)},{rng.choice(weights_option)},{rng.choice(local_restart)},{rng.choice(bound_correction)}\n"
            f.write(inst_string)
```

```python
save_cma_instances(
    "cma_bbob_ppsn_paper_train.csv", n_instances=100, dims=[5, 10, 25, 20, 25, 30]
)
save_cma_instances(
    "cma_bbob_ppsn_paper_test.csv", n_instances=100, dims=[35, 40, 45, 50, 55, 60]
)
```
