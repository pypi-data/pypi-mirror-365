```python
import pandas as pd
import matplotlib.pyplot as plt
import avinf
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv("../mtcars.csv")
df.head()
```

```python
model = ols("mpg ~ wt + hp", data=df)
model = model.fit()

av_fit = avinf.av_lm.AVLM(model, g=1)

av_fit.summary()
```

```python
# Fit a simple linear model using the mtcars dataset
model = ols("mpg ~ wt + hp", data = df)
model = model.fit()

# Convert the standard lm object to an anytime-valid avlm object with precision parameter g = 1
av_fit = avinf.av_lm.AVLM(model, g = 1)

# Calculate and print confidence intervals for the coefficients
av_fit.confint(level=0.95)
```

```python
n = 10000
alpha = 0.05
g_star = avinf.av.optimal_g(n, 5, alpha)
print("The optimal g is:", g_star)
```
