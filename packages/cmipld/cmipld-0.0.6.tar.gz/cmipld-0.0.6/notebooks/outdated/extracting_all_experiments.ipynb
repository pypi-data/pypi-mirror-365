```python
# from pprint import pprint
from pyld import jsonld
import pandas as pd
import cmipld
import json
```

```python
# get all the repositories
corpus = cmipld.fetch_all()
```

## Get all the experiments for CMIP6Plus

```python
# url = cmipld.processor.replace_prefix('cmip6plus:experiment/')
frame = {"@context":"https://wcrp-cmip.github.io/CMIP6Plus_CVs/experiment/_context_","@type":['wcrp:experiment',"cmip6plus"]}



# using the built-in JSON-LD processor
data = jsonld.frame(corpus,frame)


df = pd.DataFrame(data['@graph']).set_index('id')
df
```

### Removing the IDs 
Unless we are substituting the contents of the linked items, we do not need them in the '@id' format, so instead we can simplify the dataframe to:


```python
df = df.map(lambda x: x["id"].split('/')[-1] if (isinstance(x, dict) and 'id' in x) else x)
df.head()
```

```python
# create a template frame/context
basic = cmipld.generate_context(type='wcrp:experiment',base=None)

# print the context in json form
print(json.dumps(basic,indent=2))
```

```python
# Use the above context to list all sources. 
sources = jsonld.frame(corpus, basic)   
```

```python
# convert to a pandas dataframe
df = pd.DataFrame(sources['@graph'])

# as each column has a different prefix, standardise them
df.columns = df.columns.map(lambda x: x.split('/')[-1])
df = df.groupby(level=0, axis=1).sum()

# dereference links for easy viewing. 
df = df.map(lambda x: x["@id"].split('/')[-1] if isinstance(x, dict) else x)

df.head()
```
